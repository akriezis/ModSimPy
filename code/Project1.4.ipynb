# Will bees go exticnt in the US?

Since 1947 the number of Honey Bees colonies have been decreasing. One of the posible reasons for this sudden decline could be the increased use of insecticides in agrigulture. Honey Bees are a main natural pollinator that farmers acknowledge as good, but are inadvertly the subjects of the toxic insecticides.

To explore this question we will develop a quadratic model that predicts the number of Honey Bee colonies in the United States.

Specifically, we will add one new feature to the model
1. Insecticide Usage: The current model doesn't account for the change in insectide usage. We will use this feature to more accurately predict the number of bee colonies.

We will begin our model with the data from 1943, but begin modeling in 1947 (when the number of colonies began to decrease).

```python
population = table.Population
```

We are going to create a system that will hold the parameters of our model.

```python
t_0 = 1947;
t_end = 2200;
p_0 = population[1947];
system = System(t_0=t_0, t_end=t_end, p_0=p_0, alpha = -0.014);
```

We created an update function so we compute the population for the following year given the current population state.

```python
def update_func(pop, t, system):
    """Compute the population next year.
    
    pop: current population
    t: current year
    system: system object containing parameters of the model
    
    returns: population next year
    """
    net_growth = system.alpha  * pop
    return pop + net_growth
```

Following the update function, we created a run_simulation to establish a time series of the population.

```python
def run_simulation(system, update_func):
    """Simulate the system using any update function.
    
    system: System object
    update_func: function that computes the population next year
    
    returns: TimeSeries
    """
    results = TimeSeries()
    results[system.t_0] = system.p_0
    
    for t in linrange(system.t_0, system.t_end):
        results[t+1] = update_func(results[t], t, system)
        
    return results
```

We will now plot the imported data along side the model created by the data

```python
def plot_results(population, timeseries, title):
    """Plot the estimates and the model.
    
    census: TimeSeries of population estimates
    un: TimeSeries of population estimates
    timeseries: TimeSeries of simulation results
    title: string
    """
   
    plot(timeseries, color='gray', label='model')
    
    plot(population,':', label='Bee Population')
    decorate(xlabel='Year', 
             ylabel='Bee population (Million)',
             title='Bee population over time')
```

```python
results = run_simulation(system, update_func)
plot_results(population, results, 'Quadratic model')
```

Based on the correlation between the pesticide usage and decline in Honey Bee numbers, we will create another update function that will take into account the change in pesticide usage. We will use if statements to establish the boundaries for the death rate.

```python
def update_func2(pop, t, system):
    """Compute the population next year.
    
    pop: current population
    t: current year
    system: system object containing parameters of the model
    
    returns: population next year
    """
    if t<1972:
        net_growth = system.alpha  * pop
        return pop + net_growth
    if 1972<=t <1981:
        net_growth = system.alpha+0.004  * pop
        return pop + net_growth
    if 1981 <=t < 1997:
        net_growth = system.alpha-0.026 * pop
        return pop + net_growth
    if 1997<= t <2125:
        net_growth = system.alpha- 0.005 *pop
        return pop + net_growth
    
```

```python
results = run_simulation(system, update_func2)
plot_results(population, results, 'Quadratic model')
```

We can see that there is a negative correlation between pesticide usage and the Honey Bee population in the United States. This can clearly be seen in the last two graphs. The final graph is the data from the USDA with our model, which is adjusted based on data from the USDA. The graph is overall a good fit. The might be a discreptency in the model due to the USDA canceling the survey that collected the data of the bee populations from 1882-1885. 

The model predicts the United States will have zero Honey Bee colonies in 2125. 

This is an unfortunate reality, but will occur if humans do not change the way they interact with the environment. The model uses the current rate at which humans are using pesticides to generate the projection. If humans continue to increase the rate of pesticide usage the model will most likely change so the Honey Bees go extinct before 2125. The way to curb the plausible extnction would to stop the use of pesticide all together.
