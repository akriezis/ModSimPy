```python
## 1.36 ,0.2, 0.00136, 0.00027, 0.036, 0.33, 2, 100, 0.1  
def make_system(Γ, τ, μ, β, α, δ, σ, π, ρ):
    init = State(R=1000 , V=0.0000004 , E=0 , L=0, CD4=0)
    
    t0 = 0
    t_end =120
    
    return System(init=init, t0=t0, t_end=t_end, Γ=Γ, τ=τ, μ=μ, β=β, α=α, δ=δ, σ=σ, π=π, ρ=ρ)
```

```python
def upd_func1(state,t, system):
    unpack(system)
    r, v, e, l = state
    
    r += (Γ*τ - μ*r - β*r*v)
    return State(R=r, V=v, E=e, L=l)
```

```python
def upd_func2(state,t, system):
    unpack(system)
    r, v, e, l = state
    
    l += (ρ*β*r*v -μ*l -α*l)
    return State(R=r, V=v, E=e, L=l)
```

```python
def upd_func3(state,t, system):
    unpack(system)
    r, v, e, l = state
    
    e += (1-ρ)*β*r*v+α*l-δ*e
    return State(R=r, V=v, E=e, L=l)
```

```python
def upd_func4(state,t, system):
    unpack(system)
    r, v, e, l = state
    
    v += π*e - σ*v
    return State(R=r, V=v, E=e, L=l)
```

```python
def update_func(state,t, system):
    unpack(system)
    r, v, e, l, cd4 = state
    
    Ri= r
    Vi= v
    Ei= e
    Li= l
    
    r += ((Γ*τ) - (μ*Ri) - (β*Ri*Vi)) *0.1
    
    l += ((ρ*β*Ri*Vi) - (μ*Li) -(α*Li)) *0.1
    
    e += (((1-ρ)*β*Ri*Vi)+(α*Li)-(δ*Ei)) *0.1
    
    v += ((π*Ei) - (σ*Vi)) *0.1
    
    cd4 = 1000*(1-τ)+Ri+Li+Ei
     
    return State(R=r, V=v, E=e, L=l, CD4=cd4)
```

```python
def run_simulation(system, update_func):
    """Runs a simulation of the system.
        
    system: System object
    update_func: function that updates state
    
    returns: TimeFrame
    """
    unpack(system)
    
    frame = TimeFrame(columns=init.index)
    frame.row[t0] = init
    
    for t in linrange(t0, t_end):
        frame.row[t+1] = update_func(frame.row[t], t, system)
    
    return frame
```

```python
system = make_system(1.36 ,0.2, 0.00136, 0.00027, 0.036, 0.33, 2, 100, 0.1)
results = run_simulation(system, update_func)
```
